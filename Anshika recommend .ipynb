```python
import pandas as pd
import numpy as np
```

```python
movies_df=pd.read_csv('E:\dataset\movies.csv')
```

```python
path='E:\\dataset\\ratings.csv'
rating_df=pd.read_csv(path)
rating_df.head(5)
```

```python
ratingpivot=rating_df.pivot(index='movieId',columns='userId',values='rating').fillna(0)
```

```python
ratingpivot.head(5)
```

```python
ratingpivot.shape
```

```python
nouservoted=rating_df.groupby('movieId')['rating'].agg('count')
```

```python
nouservoted.head(5)
```

```python
nomoviesvoted=rating_df.groupby('userId')['rating'].agg('count')
```

```python
nomoviesvoted.head(5)
```

```python
finaldf=ratingpivot.loc[nouservoted[nouservoted>=20].index,:]
finaldf=finaldf.loc[:,nomoviesvoted[nomoviesvoted>=20].index]
```

```python
finaldf.head(5)
```

```python
finaldf.shape
```

```python
from scipy.sparse import csr_matrix
csrdata=csr_matrix(finaldf.values)
finaldf.reset_index(inplace=True)
finaldf.head(5)
```

```python
from sklearn.neighbors import NearestNeighbors
knn=NearestNeighbors(metric='cosine',algorithm='brute',n_neighbors=20,n_jobs=-1)
knn.fit(csrdata)
```

```python
from fuzzywuzzy import process
def movie_recommendation(moviename):
    nmovie=13
    movielist=movies_df[movies_df['title'].str.contains(moviename)]
    if len(movielist):
        #Its provide you index where movies_df title match with user moviename
        movieidx=process.extractOne(moviename,movies_df['title'])[2]
        print(movieidx)
        movieidx=movies_df['movieId'][movieidx]
        print(movieidx)
        movieidx=finaldf[finaldf['movieId']==movieidx].index[0]
        distances,indices=knn.kneighbors(csrdata[movieidx],n_neighbors=nmovie)
        indices=sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())))[1:]
        for val in indices:
            movieidx=finaldf['movieId'][val[0]]
            movieidx=movies_df[movies_df['movieId']==movieidx].index
            print('title  {0} with similarity{1}'.format(movies_df['title'][movieidx].values[0],val[1]))
    else:
        return "no movie found"
movie_recommendation('Avatar')
```
